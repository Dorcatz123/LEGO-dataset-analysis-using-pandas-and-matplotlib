### Merging (i.e., Combining) DataFrames based on a Key


```python
df_sets
```

```python
set_theme_count = pd.DataFrame({'id':df_sets["theme_id"].value_counts().index, 'set_count': df_sets["theme_id"].value_counts().values})
```

```python
set_theme_count
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id', how='inner')
```

```python
merged_df
```

```python
plt.bar(merged_df.name[:10], merged_df.loc[:,'set_count'][:10])
plt.xticks(rotation=85)
plt.xlabel('LEGO')
plt.ylabel('count')
plt.title('LEGO counts')
```

**It seems LEGO is pretty good at making themes on Star Wars, Ninjago etc**
**but it is also making products on key chains, supplementals ??**
**Has LEGO strayed from its core business or is it successfully diversifying?**
